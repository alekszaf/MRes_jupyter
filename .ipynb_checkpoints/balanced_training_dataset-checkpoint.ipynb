```python
a = pd.read_csv('C:\\Temp\\MRes\\training_data\\A_Genus_60p.csv')
b = pd.read_csv('C:\\Temp\\MRes\\training_data\\B_Genus_60p.csv')
```

```python
a.head()
```

```python
a = a.drop(columns=['X_left', 'Y_left', 'Z_left', 'freq'])
```

```python
b.head()
```

```python
b = b.drop(columns=['X', 'Y', 'freq'])
```

```python
ab = pd.concat([a, b])
```

```python
ab.head()
```

```python
ab['GENUS'].value_counts()
```

```python
ab = ab.groupby('GENUS').filter(lambda x : len(x) > 40)
```

```python
ab_3 = ab.groupby('GENUS').filter(lambda x : len(x) > 100)
```

```python
ab_3.to_csv('C://Temp//MRes//training_data//AB_3cat.csv')
```

```python
train3 = ab_3.groupby('GENUS').apply(lambda s: s.sample(100))
train4 = ab_3.groupby('GENUS').apply(lambda s: s.sample(50))

train3 = train3.reset_index(drop=True)
train4 = train3.reset_index(drop=True)
```

```python
train3.to_csv('C://Temp//MRes//training_data//AB_3cat_100.csv')
```

```python
train4.to_csv('C://Temp//MRes//training_data//AB_3cat_50.csv')
```
